```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#Read the dataset using Pandas
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
# This displays the top 5 rows of the data
dataset.head()
```

```python
# Provides some information regarding the columns in the data
dataset.info()
```

```python
# this describes the basic statistics behind the dataset used
dataset.describe()
```

```python
#Remove Position Column
dataset.drop(['Position'],inplace=True,axis=1)
```

```python
# These Plots help to explain the values and how they are scattered
plt.figure(figsize=(12,6))
sns.pairplot(dataset,x_vars=['Level'],y_vars=['Salary'],size=7,kind='scatter')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show();
```

```python
# Cooking the data
X = dataset.iloc[:,0:1].values
y = dataset.iloc[:,1].values
```

```python
#Fitting the Linear Regression to the dataset  
from sklearn.linear_model import LinearRegression  
lin_regs= LinearRegression()  
lin_regs.fit(X,y)  
```

```python
#Fitting the Polynomial regression to the dataset  
from sklearn.preprocessing import PolynomialFeatures  
poly_regs= PolynomialFeatures(degree= 4)  
x_poly= poly_regs.fit_transform(X)  
lin_reg_2 =LinearRegression()  
lin_reg_2.fit(x_poly, y)  
```

```python
#Visualizing the Linear Regression
plt.scatter(X,y,color='red')
plt.plot(X,lin_regs.predict(X))
plt.title('Linear Regresssion')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualizing the Polynomial Regression
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg_2.predict(poly_regs.fit_transform(X)))
plt.title('Polynomial Regresssion')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```
